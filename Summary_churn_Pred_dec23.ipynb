```python
cust_base_lte = pd.read_csv('pred_dec23.csv')
cust_base_lte
```

```python
cust_base_lte = cust_base_lte.rename(columns={
            'event_source' : 'EVENT_SOURCE',
            'account_num' : 'ACCOUNT_NUM' ,
            'status_threshold' : 'CHURN_PROB'  
})
cust_base_lte
```

```python
cust_base_lte = cust_base_lte[['ACCOUNT_NUM','EVENT_SOURCE','CHURN_PROB']]
cust_base_lte
```

```python
cust_base_tp = pd.read_excel('../../../Triple_Play_V2/Dec23/Churn_probability_for_thriple_play_cust_base_Dec23.xlsx')
```

```python
cust_base_tp
```

```python
cust_base_tp.isnull().sum()
```

```python
cust_base_tp = cust_base_tp.dropna(subset=['event_source'])
cust_base_tp.head()
```

```python
cust_base_tp.shape
```

```python
# copper1['EVENT_SOURCE'] = copper1['EVENT_SOURCE'].astype(str) 
cust_base_tp['event_source'] = pd.to_numeric(cust_base_tp['event_source'], errors='coerce')
cust_base_tp['event_source'] = cust_base_tp['event_source'].apply(lambda x : '{:010.0f}'.format(x) if pd.notnull(x) else '')
```

```python
cust_base_tp.head()
```

```python
cust_base_tp = cust_base_tp[cust_base_tp['churn_proba'] < 0.79]
cust_base_tp.shape
```

```python
cust_base_tp = cust_base_tp.rename(columns={
            'event_source' : 'EVENT_SOURCE',
            'churn_proba' : 'CHURN_PROB'  
})
cust_base_tp
```

```python
cust_base = pd.concat([cust_base_tp,cust_base_lte],axis=0)
```

```python
cust_base.shape
```
